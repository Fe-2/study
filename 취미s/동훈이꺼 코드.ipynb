```python
import pandas as pd
```

```python
df = pd.read_csv("두번째.csv")
```

```python
df['text'][0]
```

```python
db = pd.read_csv("12313.csv")
db['본문'] = db['본문1']
```

```python
for i in range(len(db)):
    t = str(db['본문1'][i]) + str(db['본문2'][i]) + str(db['본문3'][i])
    db['본문'][i] = t

db.drop(['본문1','본문2','본문3'],axis=1, inplace=True)
```

```python
re = []
for word in df['text']:
    i = word.lower()
    yess = []
    for j in range(len(db['본문'])):
        if str(db['본문'][j]).find(i) != -1:
            yess.append((db['카테고리'][j],db['숫자'][j]))
    re.append((i,yess))
```

```python
re
```

```python
x = []
for i in re:
    t = []
    t.append(i[0])
    p=[]
    e=[]
    f=[]
    fp=[]
    ep=[]
    for j in i[1]:
        if j[0] == 'P':
            p.append(j[1])
        if j[0] == 'E':
            e.append(j[1])
        if j[0] == 'F':
            f.append(j[1])
        if j[0] == 'F, P':
            fp.append(j[1])
        if j[0] == 'F/P':
            fp.append(j[1])
        if j[0] == 'E, P':
            ep.append(j[1])
        if j[0] == 'E/P':
            ep.append(j[1])
    t.append(f)
    t.append(e)
    t.append(p)
    t.append(fp)
    t.append(ep)
    x.append(t)
```

```python
result = pd.DataFrame(x)
result.columns = ['단어', 'F', 'E','P','FP','EP']
```

```python
result.to_csv("저장두번째.csv")
```
