```python
from rdkit import Chem
```

```python
mols =[]
with open ("drugs.smi") as fp:
    for l in fp:
        mol = Chem.MolFromSmiles(l.strip())
        mol2 = Chem.AddHs(mol)
        mols.append(mol2)
Carbon_list = []
Carbon = 0
for i in mols:
    for atom in i.GetAtoms():
        if atom.GetSymbol() == "C":
            Carbon += 1
    Carbon_list.append(Carbon)
    Carbon = 0

count = [0,0,0,0,0,0,0]
for num in Carbon_list:
    if 0<=num <=5:
        count[0] +=1
    elif 6<= num <= 10:
        count[1] +=1
    elif 11<= num <= 15:
        count[2] +=1
    elif 16<= num <= 20:
        count[3] +=1
    elif 21<= num <= 25:
        count[4] +=1
    elif 26<= num <= 30:
        count[5] +=1
    else:
        count[6] +=1
print(count)

idx = 0
for freq in count:
    print(f"{idx*5 +1} <= N <= {(idx+1)*5} : {freq}")
    idx+=1
```

```python
# 탄소갯수가 31개 이상인것들 그리기.
from rdkit.Chem import AllChem
from rdkit.Chem import Draw

large_mols =[]
for m in mols:
    n_c = 0
    for atom in m.GetAtoms():
        if atom.GetSymbol() == "C":
            n_c += 1
    if n_c <= 30:
        continue #pass
    else:
        large_mols.append(m)
        
img = Draw.MolsToGridImage(large_mols, molsPerRow=3, subImgSize = (200,200))
img.save("large_mols.png")
```

# rdMolDescriptors


```python
from rdkit.Chem import rdMolDescriptors

for m in mols:
    print(rdMolDescriptors.CalcNumHBA(m)) #hydrogen bond
```

drug.smi 파일의 분자들의 수소결합 donor와 acceptor의 갯수의 평균과 표준편차를 구해보자.

```python
import math
hba_nums = [] #acceptor
hbd_nums = [] #doner
for m in mols:
    hba_nums.append(rdMolDescriptors.CalcNumHBA(m))
    hbd_nums.append(rdMolDescriptors.CalcNumHBD(m))
    
def average(num_list):
    avg = sum(num_list) / len(num_list)
    return avg

def average_and_atd(num_list):
    avg = sum(num_list) / len(num_list)
    std = 0
    for n in num_list:
        std = std + ((n-avg)**2)
    std = std / (len(num_list)-1)
    std = math.sqrt(std)
    return avg, std
hbd_avg, hbd_std = average_and_atd(hbd_nums)
hba_avg, hba_std = average_and_atd(hba_nums)
```

```python
print(f"dor 평균/표쥰편차는: {hbd_avg:.4f} {hbd_std:.4f}")
print(f"acp 평균/표쥰편차는: {hba_avg:.4f} {hba_std:.4f}")
```
