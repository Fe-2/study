```python
all_drugs = []
with open("drugs.smi", "r") as fin:
    for l in fin:
        smi = l.strip()
        m = Chem.MolFromSmiles(smi)
        all_drugs.append(m)
```

```python
suppl = Chem.SDMolSupplier("drugs.sdf")
```

```python
for mol in suppl:
    print(mol.GetNumAtoms())
```

```python
idx = 0

for m in suppl:
    natom = m.GetNumAtoms()
    print(f"Mol No. {idx} has {natom} atoms.")
    idx += 1
```

```python
# 연습문제 : all_durg의 평균원자 갯수를 구해보자
```

```python
total = 0
for x in suppl:
    total = total + x.GetNumAtoms()

avg = total / len(suppl)
print(avg)
```
