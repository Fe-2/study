```python
import rdkit
from rdkit import Chem
```

```python
suppl = Chem.SDMolSupplier("drugs.sdf")
type(suppl)
```

```python
for mol in suppl:
    print(mol.GetNumAtoms())
```

```python
# rdkit 홈페이지 들어가면 된다.
```

```python
large_mol = []

for mol in suppl:
    num_atom = mol.GetNumAtoms()
    if num_atom >= 30:
        large_mol.append(mol)
print(len(large_mol))
```

```python
large_mol2 = [ mol for mol in suppl if mol.GetNumAtoms() >=30 ]
```

```python
print(len(large_mol2))
```
