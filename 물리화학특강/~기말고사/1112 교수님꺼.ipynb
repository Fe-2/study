# Pandas의 기본 데이터 객체 
* Pandas에서 많이 사용되는 데이터 구조는 **시리즈(Series)** 와 **데이터프레임(DataFrame)** 이 있다. 

## Series 만들기

Pandas 의 중요한 데이터 오브젝트 중 하나인 Series는 기본적으로 아래와 같이 값의 리스트를 넘겨주어 만들 수 있습니다. 

또한 값이 위치하고 있는 정보인 인덱스(index)가 Series 에 같이 저장되게 되는데요. 



```python
s = pd.Series([1, 3, 5, 6.5, 6, 8]) # Series 데이터를 초기화 한다. 
```

* 따로 전달해주지 않는 한 기본적으로 **0**부터 시작하여 **1**씩 증가하는 정수 인덱스가 사용됨을 알 수 있습니다.

```python
print(s) # 1차원 데이터. 
```

### Series 자료에서 사용할 수 있는 속성과 메소드 리스트. 

```python
type(s) # Series class
```

```python
dir(s) # Series에서 사용할 수 있는 속성과 메소드를 확인. 뭔가 많다.... 
```

```python
s.max() # 최대값 찾기 (max 메소드)
```

```python
s.min() # 최소값 찾기
```

```python
s.mean() # 평균값 찾기
```

```python
s.std()  # 표준편차.
```

```python
s.describe() # 중요한 통계값들을 자동으로 계산하여 보여줌. 
```

# DataFrame 
* 데이터프레임은 **2차원**배열이다. 엑셀을 생각하면 이해하기 쉽다. 

## DataFrame을 생성하는 방법
* 딕셔너리로 정의할 수 있다. 
* Key는 column, 데이터 개수에 맞추어서 index가 증가하면서 생성된다. 

```python
df2 = pd.DataFrame({'A': (1.0, 2.0, 3.0, 4.0),
                    'Date': [pd.Timestamp('20130102'), pd.Timestamp('20130103'), pd.Timestamp('20130104'), pd.Timestamp('20130105')],
                    'C': pd.Series((1, 3.0, 6.0, 9.0), dtype='float32'),
                    'D': np.array([4,5,6,9], dtype='int32'),
                    'E': pd.Categorical(['test', 'train', 'test', 'train']),
                    'F': 'foo'})
```

```python
print(df2)
```

```python
dir(df2) # DataFrame에서 사용할 수 있는 속성 및 메쏘드 리스트.... 좀 많죠? 
```

```python
type(df2)
```

# 전체 데이터 셋 분리!
* 전체 데이터를 학습용 데이터와 테스트용 데이터로 분리해야 한다!

```python
import sklearn
import sklearn.model_selection
iris_df
iris_df.describe()
```

```python
train_set, test_set = sklearn.model_selection.train_test_split(iris_df, test_size=0.2, random_state=1234) # test_size의 기본값은 0.25 (25%)
```

```python
train_set.describe()
```

```python
test_set.describe()
```

```python
train_set
```

## 데이터 정렬

```python
train_set.sort_index(axis=0) # 인덱스를 이용해서 오름차순!
```

```python
train_set.sort_index(axis=0, ascending=False) # 내림차순. 
```
