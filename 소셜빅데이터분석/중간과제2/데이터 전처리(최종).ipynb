```python
import pandas as pd
```

```python
경향 = pd.read_csv('경향신문_인공지능(20100101~20201231).csv')
동아 = pd.read_csv('동아신문_%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5(2020-12-31~2010-01-01).csv')
#경향과 동아 pd로 불러와주고,
경향['Species'] = "경향"
동아['Species'] = "동아"
# 딱히 생각나는 단어가 없어서... "Species" 라고 붙였습니다. Kinds가 나았을까요
경향['Date'] = 경향['Date'].str.replace('.', '')
동아['Date'] = 동아['Date'].str.replace('-', '')
# date부분의 형식이 경향과 동아가 달라서 형식을 동일하게 YYYYMMDD로 나타내었습니다.
```

```python
전체 = pd.concat([경향, 동아])
# 두개의 pd를 합쳐주고,
```

```python
전체.to_csv('전처리전.csv', encoding = 'utf-8-sig', index_label = False, index = False)
# 전처리전.csv로 저장해줍니다. 
# 백업의 용도도 있고,
# index를 초기화 하기위해서 이렇게 했습니다.
```

```python
df = pd.read_csv('전처리전.csv')
# 전체 데이터를 새로 불러와줍니다.
# index가 초기화되어 0부터 14105까지인 것을 확인할수 있습니다.
df
```

```python
df["Content"] = df["Content"].str.replace(pat=r'[^\w]', repl=r' ', regex=True)
df["Title"] = df["Title"].str.replace(pat=r'[^\w]', repl=r' ', regex=True)
# 마지막으로미뤄두었던 모든 특수문자를 제거해줍니다. 
# ■, \n, ^ , . 전부 없앴습니다.
# 혹시 . 이나, , , 혹은 () 등은 필요하다면 전처리방법을 바꾸면 됩니다.(백업)
df
#전처리가 잘 된것을 확인할 수 있습니다.
```

```python
# 저장해줍니다. 이상입니다.
df.to_csv('최종.csv', encoding = 'utf-8-sig', index_label = False, index = False)
```
