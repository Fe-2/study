```python
import pandas as pd
#import steamreviews  # https://pypi.org/project/steamreviews/
import download_reviews
import time
from datetime import datetime
from datetime import date
import os

import matplotlib.pyplot as plt
```

```python
l = []

app_id = 1599340

request_params = dict()                   
request_params['language'] = 'english' 
#request_params['day_range'] = 171
request_params['review_type'] = 'positive'

#review_dictP, query_count = steamreviews.download_reviews_for_app_id(app_id,chosen_request_params=request_params,verbose= True)
review_dictP, query_count, cursor, num_reviews_X = download_reviews.download_reviews_for_app_id(app_id,chosen_request_params=request_params,verbose = True)

for i in review_dictP['reviews']:
    ID = review_dictP['reviews'][i]['author']['steamid']
    review = review_dictP['reviews'][i]['review']
    steam_purchase =  review_dictP['reviews'][i]['steam_purchase']
    playtime = review_dictP['reviews'][i]['author']['playtime_at_review']
    language = review_dictP['reviews'][i]['language']
    create_timestamp = review_dictP['reviews'][i]['timestamp_created']
    update_timestamp = review_dictP['reviews'][i]['timestamp_updated']
    l.append((ID,review,steam_purchase,playtime,language,datetime.fromtimestamp(int(create_timestamp)),datetime.fromtimestamp(int(update_timestamp)),1,cursor))

os.remove('./data/review_1599340.json')
print(f'긍정적인 review 수는 현재 {len(l)}개')  
print('\n==============\n')  

while True:
    skajwl = len(l)%100
    if skajwl != 0:
        break
        
    time.sleep(310)
    print(f"Start_cursor = {cursor}")
    beforenum = len(l)
    review_dictP, query_count, cursor_Y, num_reviews = download_reviews.download_reviews_for_app_id(app_id,chosen_request_params=request_params,start_cursor=cursor,verbose = True)

    for i in review_dictP['reviews']:
        ID = review_dictP['reviews'][i]['author']['steamid']
        review = review_dictP['reviews'][i]['review']
        steam_purchase =  review_dictP['reviews'][i]['steam_purchase']
        playtime = review_dictP['reviews'][i]['author']['playtime_at_review']
        language = review_dictP['reviews'][i]['language']
        create_timestamp = review_dictP['reviews'][i]['timestamp_created']
        update_timestamp = review_dictP['reviews'][i]['timestamp_updated']
        l.append((ID,review,steam_purchase,playtime,language,datetime.fromtimestamp(int(create_timestamp)),datetime.fromtimestamp(int(update_timestamp)),1,cursor))

    os.remove('./data/review_1599340.json')
    print(f'긍정적인 review 수는 현재 {len(l)}개') 
    print('\n==============\n')   
    if not beforenum == len(l):
        cursor = cursor_Y
        
#============================================#    
#============================================#    
#============================================#    

m = []

request_params['review_type'] = 'negative'


review_dictN, query_count, cursor, num_reviews_X = download_reviews.download_reviews_for_app_id(app_id,chosen_request_params=request_params,verbose = True)

for i in review_dictN['reviews']:
    ID = review_dictN['reviews'][i]['author']['steamid']
    review = review_dictN['reviews'][i]['review']
    steam_purchase =  review_dictN['reviews'][i]['steam_purchase']
    playtime = review_dictN['reviews'][i]['author']['playtime_at_review']
    language = review_dictN['reviews'][i]['language']
    create_timestamp = review_dictN['reviews'][i]['timestamp_created']
    update_timestamp = review_dictN['reviews'][i]['timestamp_updated']
    m.append((ID,review,steam_purchase,playtime,language,datetime.fromtimestamp(int(create_timestamp)),datetime.fromtimestamp(int(update_timestamp)),0,cursor))

os.remove('./data/review_1599340.json')
print(f'부정적인 review 수는 현재 {len(m)}개')
print('\n==============\n') 

while True:
    skajwl = len(m)%100
    if skajwl != 0:
        break
        
    time.sleep(60)
    print(f"Start_cursor = {cursor}")
    beforenum = len(m)
    review_dictN, query_count, cursor_Y, num_reviews = download_reviews.download_reviews_for_app_id(app_id,chosen_request_params=request_params,start_cursor=cursor,verbose = True)

    for i in review_dictN['reviews']:
        ID = review_dictN['reviews'][i]['author']['steamid']
        review = review_dictN['reviews'][i]['review']
        steam_purchase =  review_dictN['reviews'][i]['steam_purchase']
        playtime = review_dictN['reviews'][i]['author']['playtime_at_review']
        language = review_dictN['reviews'][i]['language']
        create_timestamp = review_dictN['reviews'][i]['timestamp_created']
        update_timestamp = review_dictN['reviews'][i]['timestamp_updated']
        m.append((ID,review,steam_purchase,playtime,language,datetime.fromtimestamp(int(create_timestamp)),datetime.fromtimestamp(int(update_timestamp)),0,cursor))

    os.remove('./data/review_1599340.json')
    print(f'부정적인 review 수는 현재 {len(m)}개')
    print('\n==============\n')
    if not beforenum == len(m):
        cursor = cursor_Y
    
    
```

## 요거는 날짜빵꾸난거 확인

```python
dfP = pd.DataFrame(l)
dfN = pd.DataFrame(m)
df = pd.merge(dfP,dfN, how= 'outer')
df.columns = ['UserID', 'Review','Purchase','Playtime','language','Create-Date','Up-Date','Positive=1','Cursor']
df['Raw'] = df['Review']
df['Review'] = df['Review'].str.lower()
df['Review'] = df['Review'].str.replace(pat=r'[\s]', repl= r' ')
df['Review'] = df['Review'].str.replace(pat=r'\[[^)]*\]', repl= r' ')
df.to_csv('scraping_data.csv', index = False)
df = pd.read_csv(f'scraping_data.csv')
lenscraping = len(df)
df2 = pd.read_csv('raw_data.csv')
lenraw1 = len(df2)
df =  pd.concat([df,df2])
lenraw2 = len(df)
df = df.sort_values(by=['Up-Date'], ascending=False)
df = df.drop_duplicates('UserID')
lenraw3 = len(df)

print(f"scrapping_data = {lenscraping} 개\n")

print(f"추가 = {(lenraw3 - lenraw1)} 개\n")

print(f"총 = {lenraw3} 개")

df.to_csv('raw_data.csv', index = False)
```

```python
df = pd.read_csv('raw_data.csv')
x = len(df)

df['Up-Date'] = df['Up-Date'].str[:10]
df['Create-Date'] = df['Create-Date'].str[:10]
df['Up-Date'] = pd.to_datetime(df['Up-Date'], format='%Y-%m-%d',errors='coerce').dt.date
df['Create-Date'] = pd.to_datetime(df['Create-Date'], format='%Y-%m-%d',errors='coerce').dt.date
df.drop(df.loc[df['Create-Date'] < date(2022, 2, 11) ].index, inplace=True)
df.drop(df.loc[df['Up-Date'] < date(2022, 2, 11) ].index, inplace=True)
dfP = df[df['Positive=1'] == 1]
dfN = df[df['Positive=1'] == 0]
y = len(df)

print(f'날짜데이터가 이상한 {x-y} 개 삭제\n')

print(f'최종데이터 {y}개')

df.to_csv('.csv', index = False)
```

# 전체 시각화

```python
numP = len(dfP)
numN = len(dfN)

plt.figure(figsize=(15,10))
ratio = [numP, numN]
labels = ['Recommend', 'Not Recommend']
colors = ['silver', 'gold', 'whitesmoke', 'lightgray']
explode = [0.025, 0.025]
#plt.title('Recommend/Not Recommend Freq', fontsize = 35)
plt.pie(ratio, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, colors= ['#86c5e5','#ff9999'],shadow=True,textprops={'fontsize': 35}, labels=labels)
plt.axis('equal')
plt.savefig('./picture/pie.png')
plt.show()
```

```python
Po = []
date1 = []
date2 = []
for i in dfP['Up-Date'].value_counts().sort_index():
    Po.append(i)
for i in dfP['Up-Date'].value_counts().sort_index().index:
    date1.append(i.strftime("%m/%d"))
    date2.append(i)
    
freqP = pd.DataFrame(list(zip(date1,Po)), columns = ['Up-Date','Freq'])

ax = freqP.plot(x='Up-Date',y='Freq',kind='bar', title="Recommend", rot=0,fontsize=10,figsize=(60,10),color='#86c5e5')
plt.title('Recommend', fontsize=30)
#plt.savefig('./picture/Recommend_freq.png')
```

```python
Ne = []
date1 = []
date3 = []
for i in dfN['Up-Date'].value_counts().sort_index():
    Ne.append(i)
for i in dfN['Up-Date'].value_counts().sort_index().index:
    date1.append(i.strftime("%m/%d"))
    date3.append(i)
    
freqN = pd.DataFrame(list(zip(date1,Ne)), columns = ['Up-Date','Freq'])

ax = freqN.plot(x='Up-Date',y='Freq',kind='bar', title="Not Recommend", rot=0,fontsize=10,figsize=(60,10),color='#ff9999')
plt.title('Not Recommend', fontsize=30)
#plt.savefig('./picture/Not Recommend_freq.png')
```

```python
date4 = sorted(list(set(date2 + date3)))


plt.rcParams["figure.figsize"] = (30,10)
plt.rcParams.update({'font.size': 22})
plt.rcParams['lines.linewidth'] = 3
plt.plot(date4,freqP['Freq'],color='#86c5e5')
plt.plot(date4,freqN['Freq'],color='#ff9999')
plt.legend(['Recommend','Not Recommend'])
plt.title('Recommend/Not Recommend Freq', fontsize=45)

plt.savefig('./picture/line.png')
```
