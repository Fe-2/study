```python
from selenium import webdriver
options = webdriver.ChromeOptions() 
options.add_argument('--headless') 
options.add_argument('--no-sandbox') 
options.add_argument('--disable-dev-shm-usage') 
from bs4 import BeautifulSoup
import pandas as pd
pd.set_option('display.max_colwidth', None)
import urllib.request
import urllib.parse
from urllib.parse import quote
import time
import datetime
```

```python
starttime = time.time()
date = []
number = []
firstURL = 'https://gnews.gg.go.kr/briefing/brief_covid19.do?page=1&search=0&keyword=&search_type=&CU_CODE1=NA06&CU_CODE2=NB39' 
req = urllib.request.Request(firstURL, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.75 Safari/537.36'})        
source_code_from_url = urllib.request.urlopen(req)
firstsoup = BeautifulSoup(source_code_from_url, 'lxml', from_encoding='utf-8')
num = int(firstsoup.select('#chk-table > tbody > tr:nth-child(1) > td:nth-child(1)')[0].get_text())
page_num = int(num / 10) +1

for i in range(1,page_num+1):
    URL = f"https://gnews.gg.go.kr/briefing/brief_covid19.do?page={i}&search=0&keyword=&search_type=&CU_CODE1=NA06&CU_CODE2=NB39"
    req = urllib.request.Request(URL, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.75 Safari/537.36'})        
    source_code_from_url = urllib.request.urlopen(req)
    soup = BeautifulSoup(source_code_from_url, 'lxml', from_encoding='utf-8')
    try:
        for j in range(1,11):
            day = soup.select(f'#chk-table > tbody > tr:nth-child({j}) > td:nth-child(3)')[0].get_text()
            num = soup.select(f'#chk-table > tbody > tr:nth-child({j}) > td:nth-child(1)')[0].get_text()
            date.append(day)
            number.append(num)
    except:
        pass
    
    print(f'{i} 페이지완료')
    
    if i % 100 == 0:
        print('쉬는중입니다.')
        print(f'{len(date)}명 집계완료')
        time.sleep(300)
              
runtime = time.time() - starttime
              
print(f'끝! 걸린시간 = {runtime}초')
print(f'총 {len(date)}명 집계완료')             
```

```python
df = pd.DataFrame(list(zip(date,number)), columns =['Date','경기도연번'])
df.to_csv('코로나확진자날짜.csv', encoding = 'utf-8-sig', index_label = False, index = False)
```
