```python
import pandas as pd
import datetime
df = pd.read_csv('서울시 코로나19 확진자 현황.csv')
```

```python
df
```

```python
date = []
확진일 = []
일일확진자수 = []
for x in df['확진일']:
    y = str(datetime.datetime.strptime(x, "%Y-%m-%d"))[:10]
    확진일.append(x)
    date.append(y)
```

```python
date = set(date)
date = list(date)
date.sort()

for i in range(0,len(date)):
    확진자 = 확진일.count(date[i])
    일일확진자수.append(확진자)
```

```python
savedf = pd.DataFrame(list(zip(date,일일확진자수)), columns =['date','일일확진자수'])
savedf.to_csv('서울시 코로나일일확진자.csv', encoding = 'utf-8-sig', index_label = False, index = False)
```
