```python
import pandas as pd
```

```python
for j in range(202101,202103):
    df = pd.read_csv(f'CARD_SUBWAY_MONTH_{j}.csv' )
    datelist = []
    subway = []
    num = j*100 +1
    for i in range(num,num+32):
        df2 = df.loc[df['사용일자'] == i, '승차총승객수']
        a = df2.sum()
        datelist.append(i)
        subway.append(a)

    savedf = pd.DataFrame(list(zip(datelist,subway)), columns =['date','subway'])
    savedf = savedf[savedf.subway != 0]
    savedf.to_csv(f'subway_{j}.csv', encoding = 'utf-8-sig', index_label = False, index = False)
```
