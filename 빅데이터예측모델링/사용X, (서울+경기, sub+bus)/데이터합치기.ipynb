```python
import pandas as pd
import datetime
from datetime import timedelta
```

```python
서울 = pd.read_csv('서울시코로나일일_최종.csv')
경기 = pd.read_csv('경기도코로나일일_최종.csv')
버스 = pd.read_csv('bus_최종.csv')
지하철 = pd.read_csv('subway_최종.csv')
서울_전날 = pd.read_csv('서울시코로나일일_최종.csv')
경기_전날 = pd.read_csv('경기도코로나일일_최종.csv')

for i in range(0,len(버스['date'])):
    버스['date'][i] = str(datetime.datetime.strptime(str(버스['date'][i]),'%Y%m%d'))[:10]
    지하철['date'][i] = str(datetime.datetime.strptime(str(지하철['date'][i]),'%Y%m%d'))[:10]
    
for i in range(0,len(경기['date'])):
    경기['date'][i] = datetime.datetime.strptime(경기['date'][i],'%Y-%m-%d')
    경기_전날['date'][i] = 경기['date'][i]+ timedelta(days=1)
    경기_전날['date'][i] = str(경기_전날['date'][i])[:10]

for i in range(0,len(서울['date'])):
    서울['date'][i] = datetime.datetime.strptime(서울['date'][i],'%Y-%m-%d')
    서울_전날['date'][i] = 서울['date'][i]+ timedelta(days=1)
    서울_전날['date'][i] = str(서울_전날['date'][i])[:10]

서울_전날.rename(columns = {'서울일일확진자수' : '서울전날확진자수'}, inplace = True)
경기_전날.rename(columns = {'경기일일확진자수' : '경기전날확진자수'}, inplace = True)
```

```python
df = pd.merge(버스, 지하철)
df = pd.merge(df, 경기_전날, on ='date',how='outer')
df = pd.merge(df, 서울_전날, on ='date',how='outer')
df = df.dropna(subset=['bus'])
df = df.fillna(0)
df['수도권전날확진자수'] = df['경기전날확진자수'] + df['서울전날확진자수']
df['수도권누적확진자수'] = 0
df['주말'] = 0
k=0

for i in range(0,len(df['수도권누적확진자수'])):
    k += df['경기전날확진자수'][i]
    k += df['서울전날확진자수'][i]
    df['수도권누적확진자수'][i] = k
    
for i in range(0,len(df['date'])):
    x = datetime.datetime.strptime(df['date'][i],'%Y-%m-%d')
    요일 = x.weekday()
    if 요일 == 5:
        df['주말'][i] = 1
    elif 요일 == 6:
        df['주말'][i] = 1
    else:
        df['주말'][i] = 0
```
